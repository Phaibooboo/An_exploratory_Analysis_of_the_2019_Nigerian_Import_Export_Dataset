# Data Cleaning
Here, we deal with the issues highlighted during the data assessment. Before we begin however, we need to create a new dataframe that we would work on so as not to lose any data from the original data

```python
#making a copy of the dataset which will be cleaned to preserve the original
imports_clean = imports.copy()
```

### 1.  Addressing first issue of inaccurate data type...

**Define : Inaccurate data type (APP_YEAR)**

To-do: Convert  'APP_YEAR' datatype from float to datetime


**Code**

```python
#converting APP_YEAR column to datetime
imports_clean['APP_YEAR'] = pd.to_datetime(imports_clean['APP_YEAR'])
```

**Test**

```python
#checking the datatype of APP_YEAR column
imports_clean['APP_YEAR'].info()
```

### 2.  The next issue of duplicate columns

**Define : Missing values (Bank code, Bank name)**

To-do: Drop these columns as we have already identified two columns which holds the needed complete information; BANK_CODE and BNK_NAM.

**Code**

```python
#dropping the duplicate columns
imports_clean = imports_clean.drop(['Bank Code', 'Bank name'], axis = 1)
```

**Test**

```python
#Confirm that columns have been dropped
imports_clean.columns
```

### 3.  Now the issue of incorrectly spelt column name

**Define : Nondescriptive/Incorrectly spelt column name (BNK_NAM)**

To-do: Rename BNK_NAM column

**Code**

```python
#renaming the BNK_NAM column
imports_clean.rename(columns = {'BNK_NAM': 'BANK_NAME'}, inplace = True)
```

**Test**

```python
#confirm renaming
imports_clean.info()
```

### 4.  Dealing with the null rows

**Define : Presence of null rows**

To-Do : Drop null rows (or fill if necessary

```python
#checking for the number of null cells in each column
imports_clean.isnull().sum()
```

Each column has at least 1 null cell. IMPORTER_ADDRESS seems to have a lot more with 138 null cells. Applying the dropna method will drop 138 cells and losing that amount of data can be avoided by filling the nulls in the IMPORTER_ADDRESS with 'Not provided' and then applying the dropna method will drop 3 rows instead

**Code 1**

```python
#filling the nulls in the 'IMPORTER_ADDRESS with 'Not provided'
imports_clean.fillna({'IMPORTER_EMAIL': 'Not provided'}, inplace = True)
```

**Test 1**

```python
#checking the number of nulls in each column again to confirm the fill
imports_clean.isnull().sum()
```

**Code 2**

```python
#dropping the null rows (3 rows)
imports_clean.dropna(inplace =True)
```

**Test 2**

```python
#confirming the drop
imports_clean.isnull().sum()
```

### 5. Splitting columns into smaller useful columns

**Define : AUTHORIZED_DEALER_DATE column can be divided into distinct columns (day, month)**

To-do: Divide AUTHORIZED_DEALER_DATE column into 2 distinct columns (day, month)

**Code**

```python
#splitting AUTHORIZED_DEALER_DATE column into two distinct columns (day, month)
imports_clean['Month'] = imports_clean['AUTHORIZED_DEALER_DATE'].dt.month
imports_clean['Day'] = imports_clean['AUTHORIZED_DEALER_DATE'].dt.day
```

**Test**

```python
#checking for correctly created columns
imports_clean.columns
```

### 5. What months experience increased export traffic?

```python
#Plotting a bar chart showing the volume of exports per month
fig, sub = plt.subplots(1,1)
sub = imports_clean.groupby("Month")['S/N'].count().plot(kind = 'bar')

plt.xlabel("Months",fontsize = 18)
plt.ylabel("Export Volume",fontsize = 18)
sub.set_xticks(range(0,12))
sub.set_xticklabels(['January', 'Febuary', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.xticks(rotation = 50)
plt.title("Volume Imports Per Month\n",fontsize = 18)
```

Most of the exports arrived the country in the month of July. This is a possible explanation as to why dollar spiked so high in July as seen in the previous chart. 

### 6. Who are the top importers?

We'll answer this question based on frequency of imports and amount spent on imports.

For the frequency of imports

```python
#plotting the top importing companies 
top_importers=imports_clean.groupby('APPLICANT_NAME')['APPLICANT_NAME'].count().sort_values(ascending=False).head(10) #slicing to get the top 10
top_importers.plot(kind='barh').invert_yaxis()


plt.ylabel("Company",fontsize = 20)
plt.xlabel("\nVolume of import",fontsize = 20)
plt.xticks(rotation = 0)
plt.title("Top Importers (Based on Frequency of Imports)\n",fontsize = 20)
```

For top importers based on amount spent on imports

```python
top_importers2 = imports_clean.groupby('APPLICANT_NAME')['USD EQUIV'].sum().sort_values(ascending=False).head(10)
top_importers2.plot(kind='barh').invert_yaxis()


plt.ylabel("Company",fontsize = 20)
plt.xlabel("\nVolume of import",fontsize = 20)
plt.xticks(rotation = 0)
plt.title("Top Importers(Based on amount spent on imports)\n",fontsize = 20)
```

Nestle is the top importer based on frequency of import, Crown flour mills is the top importer based on the amount spent on imports.

### 7. What is the most and least expensive item imported?

```python
fig, sub = plt.subplots(1,2, figsize = (15,5))
fig.subplots_adjust(wspace = 1.0)

#for the most expensive items
imports_clean.groupby('ITEM')['FOB Value Naira'].mean().sort_values(ascending = False).head(10).plot(kind = 'barh', ax=sub[0])

sub[0].set_ylabel("Item",fontsize = 20)
sub[0].set_xlabel("Amount (in billions)",fontsize = 20)
sub[0].set_yticklabels(['LAMINATING FLOOR MACHINE', 
                        'PARTS FOR FERTILIZER PROJECT',
                        'SPARE PARTS',
                        'MACHINERY AND ACCESSORIES',
                        'ROUND FLOOR PALLET',
                        'HYGENIC PRODUCTS',
                        'BLADES AND RAZOR',
                        'SPARE PARTS',
                        'SINOTRUCK HOWO TRACTOR',
                        'SPARE PARTS'])
sub[0].set_title("Top 10 items based on Price\n",fontsize = 20)

#for the least expensive products
imports_clean.groupby('ITEM')['FOB Value Naira'].mean().sort_values()[1:11].plot(kind = 'barh', ax=sub[1])

sub[1].set_ylabel("Item",fontsize = 20)
sub[1].set_xlabel("Amount",fontsize = 20)
sub[1].set_yticklabels(['FLASK', 
                        'MEDICAL DIAGNOSTIC REAGENT',
                        'SODIUM ASCORBATE',
                        'SPARE PARTS',
                        'LG SPARE PARTS',
                        'ODORIFEROUS SUBSTANCES',
                        'FLAVOURING ESSENCE',
                        'SPARE PARTS',
                        'CALCUIM CARBONATE',
                        'KITCHEN AND HOUSEKEEPING PRODUCTS'])
sub[1].set_title("Least 10 items based on Price\n",fontsize = 20)
```

The most expensive item that was imported in 2019 is the laminating flooring machine. It is used for laminating floor. The least expensive item that was imported is flask.
